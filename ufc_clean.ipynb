```python
# Import Dependencies
import pandas as pd
import datetime as dt
import scipy.stats as st
import re
```

```python
# Define File Path
history = "ufc_history.csv"

# # Colab URL
# history = "https://raw.githubusercontent.com/GregoryRash6/Project-3/master/ufc_history.csv"

# Read CSV
ufc_history = pd.read_csv(history)

# Reverse CSV Order
ufc_history = ufc_history.iloc[::-1]
```

```python
# Define File Path
current = "ufc_current.csv"

# # Colab URL
# current = "https://raw.githubusercontent.com/GregoryRash6/Project-3/master/ufc_current.csv"

# Read CSV
ufc_current = pd.read_csv(current)

# Reverse CSV Order
ufc_current = ufc_current.iloc[::-1]
```

```python
# Assign Date Column to a Variable
all_dates_history = ufc_history["date"]
all_dates_current = ufc_current["date"]

# Convert All Dates to a Datetime 
[dt.datetime.strptime(x, "%Y-%m-%d") for x in all_dates_history]

# Replace Date Column with Datetime Values
ufc_history["date"] = pd.to_datetime(all_dates_history)

# Convert All Dates to a Datetime 
[dt.datetime.strptime(x, "%m/%d/%Y") for x in all_dates_current]

# Replace Date Column with Datetime Values
ufc_current["date"] = pd.to_datetime(all_dates_current)
```

```python
# Remove Overlapping Data
ufc_current = ufc_current[ufc_current["date"] > "2019-06-08"]
```

```python
# Set Columns to Keep
common_cols = ['B_Stance', 'R_win_by_Decision_Unanimous', 'R_win_by_KO/TKO', 'R_wins', 'B_avg_SIG_STR_pct', 'R_avg_TD_pct', 'R_age', 'R_avg_SIG_STR_landed', 'B_losses', 'B_win_by_Submission', 'B_draw', 'B_avg_TD_pct', 'B_longest_win_streak', 'R_fighter', 'R_win_by_TKO_Doctor_Stoppage', 'R_Stance', 'B_Weight_lbs', 'B_total_title_bouts', 'B_win_by_Decision_Unanimous', 'B_win_by_Decision_Split', 'R_avg_TD_landed', 'B_current_win_streak', 'B_avg_TD_landed', 'date', 'R_total_rounds_fought', 'B_win_by_Decision_Majority', 'no_of_rounds', 'B_age', 'B_avg_SUB_ATT', 'R_total_title_bouts', 'R_win_by_Decision_Split', 'R_losses', 'B_current_lose_streak', 'B_fighter', 'B_Height_cms', 'R_Reach_cms', 'location', 'R_avg_SIG_STR_pct', 'B_total_rounds_fought', 'R_draw', 'B_avg_SIG_STR_landed', 'B_win_by_KO/TKO', 'B_Reach_cms', 'R_win_by_Submission', 'R_longest_win_streak', 'R_win_by_Decision_Majority', 'R_current_win_streak', 'R_Height_cms', 'Winner', 'R_avg_SUB_ATT', 'weight_class', 'B_win_by_TKO_Doctor_Stoppage', 'R_current_lose_streak', 'R_Weight_lbs', 'title_bout', 'B_wins']

# Make a Copy, Keeping Columns Listed Above
history_transformed = ufc_history[common_cols].copy()

# Make a Copy, Keeping Columns Listed Above
current_transformed = ufc_current[common_cols].copy()
```

```python
# Empty List
weight_bucket = []

# For List to Bucket Weight Classes
for x in range(len(ufc_complete)):
    if ufc_complete["weight_class"][x] == "Flyweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Bantamweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Featherweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Lightweight":
        weight_bucket.append("Middle")
    elif ufc_complete["weight_class"][x] == "Welterweight":
        weight_bucket.append("Middle")
    elif ufc_complete["weight_class"][x] == "Middleweight":
        weight_bucket.append("Heavy")
    elif ufc_complete["weight_class"][x] == "Light Heavyweight":
        weight_bucket.append("Heavy")
    elif ufc_complete["weight_class"][x] == "Heavyweight":
        weight_bucket.append("Heavy")
    else:
        print(ufc_complete["weight_class"][x])
```

```python
# Add Weight Buckets to Dataframe
ufc_complete["weight_buckets"] = weight_bucket
```

```python
# Create Empty List
blue_win = []

# For List Hot or Not
for x in range(len(ufc_complete)):
    if ufc_complete["B_current_win_streak"][x] == 0:
        blue_win.append(0)
    else:
        blue_win.append(1)

# Add Hot or Not to Dataframe
ufc_complete["B_hot_or_not"] = blue_win

# Create Empty List
red_win = []

# For List Hot or Not
for x in range(len(ufc_complete)):
    if ufc_complete["R_current_win_streak"][x] == 0:
        red_win.append(0)
    else:
        red_win.append(1)

# Add Hot or Not to Dataframe
ufc_complete["R_hot_or_not"] = red_win
```

```python
# Assign Columns to Variables
r_height_cm = ufc_complete["R_Height_cms"]
b_height_cm = ufc_complete["B_Height_cms"]
r_weight_lbs = ufc_complete["R_Weight_lbs"]
b_weight_lbs = ufc_complete["R_Weight_lbs"]

# Create Empty Lists
r_meters = []
b_meters = []
r_kilograms = []
b_kilograms = []
r_bmi = []
b_bmi = []

# Convert to Meters
for x in range(len(r_height_cm)):
    r_meters.append(r_height_cm[x] / 100)

# Convert to Meters
for x in range(len(b_height_cm)):
    b_meters.append(b_height_cm[x] / 100)

# Convert to Kilograms
for x in range(len(r_weight_lbs)):
    r_kilograms.append(r_weight_lbs[x] / 2.20462)
    
# Convert to Kilograms
for x in range(len(b_weight_lbs)):
    b_kilograms.append(b_weight_lbs[x] / 2.20462)
    
# Calculate BMI
for x in range(len(r_kilograms)):
    r_bmi.append(r_kilograms[x]/(r_meters[x] ** 2))
    
# Calculate BMI
for x in range(len(b_kilograms)):
    b_bmi.append(b_kilograms[x]/(b_meters[x] ** 2))

# Create Dataframe
bmi = pd.DataFrame(list(zip(b_bmi, r_bmi)),
              columns=['B_BMI','R_BMI'])

# Show Data
bmi
```

```python
# Merge Data
ufc_complete = pd.merge(ufc_complete, bmi, how="outer", left_index=True, right_index=True)
```
