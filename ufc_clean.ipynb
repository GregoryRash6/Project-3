```python
# Import Dependencies
import pandas as pd
import datetime as dt
import scipy.stats as st
```

```python
# Define File Path
history = "ufc_history.csv"

# Read CSV
ufc_history = pd.read_csv(history)

# Reverse CSV Order
ufc_history = ufc_history.iloc[::-1]
```

```python
# Define File Path
current = "ufc_current.csv"

# Read CSV
ufc_current = pd.read_csv(current)

# Reverse CSV Order
ufc_current = ufc_current.iloc[::-1]
```

```python
# Assign Date Column to a Variable
all_dates_history = ufc_history["date"]
all_dates_current = ufc_current["date"]

# Convert All Dates to a Datetime 
[dt.datetime.strptime(x, "%Y-%m-%d") for x in all_dates_history]

# Replace Date Column with Datetime Values
ufc_history["date"] = pd.to_datetime(all_dates_history)

# Convert All Dates to a Datetime 
[dt.datetime.strptime(x, "%m/%d/%Y") for x in all_dates_current]

# Replace Date Column with Datetime Values
ufc_current["date"] = pd.to_datetime(all_dates_current)
```

```python
ufc_current = ufc_current[ufc_current["date"] > "2019-06-08"]
```

```python
# Set Columns to Keep
history_cols = ["B_fighter", "R_fighter", "date", "location", "Winner", "title_bout", "weight_class", "no_of_rounds", "B_current_lose_streak", "B_current_win_streak", "R_current_lose_streak", "R_current_win_streak", "B_total_rounds_fought", "R_total_rounds_fought", "B_wins", "R_wins", "B_Stance", "R_Stance", "B_Height_cms", "R_Height_cms", "B_Reach_cms", "R_Reach_cms", "B_Weight_lbs", "R_Weight_lbs", "B_age", "R_age"]

# Make a Copy, Keeping Columns Listed Above
history_transformed = ufc_history[history_cols].copy()
```

```python
# Set Columns to Keep
current_cols = ["B_fighter", "R_fighter", "date", "location", "Winner", "title_bout", "weight_class", "no_of_rounds", "B_current_lose_streak", "B_current_win_streak", "R_current_lose_streak", "R_current_win_streak", "B_total_rounds_fought", "R_total_rounds_fought", "B_wins", "R_wins", "B_Stance", "R_Stance", "B_Height_cms", "R_Height_cms", "B_Reach_cms", "R_Reach_cms", "B_Weight_lbs", "R_Weight_lbs", "B_age", "R_age"]

# Make a Copy, Keeping Columns Listed Above
current_transformed = ufc_current[current_cols].copy()
```

```python
frames = [history_transformed, current_transformed]

ufc_complete = pd.concat(frames)

ufc_complete
```

```python
ufc_complete = ufc_complete.dropna()
```

```python
ufc_complete = ufc_complete[ufc_complete["weight_class"] != "Women's Strawweight"]
ufc_complete = ufc_complete[ufc_complete["weight_class"] != "Women's Bantamweight"]
ufc_complete = ufc_complete[ufc_complete["weight_class"] != "Women's Flyweight"]
ufc_complete = ufc_complete[ufc_complete["weight_class"] != "Women's Featherweight"]
ufc_complete = ufc_complete[ufc_complete["weight_class"] != "Catch Weight"]
```

```python
# Get names of indexes for which column Age has value 30
indexNames = ufc_complete[ufc_complete['Winner'] == "Draw" ].index
# Delete these row indexes from dataFrame
ufc_complete.drop(indexNames , inplace=True)
```

```python
ufc_complete = ufc_complete.reset_index(drop=True)
```

```python
len(ufc_complete["location"].unique())
```

```python
# Define File Path
elevation = "ufc_elevation.csv"

# Read CSV
ufc_elevation = pd.read_csv(elevation)
```

```python
ufc_complete = pd.merge(ufc_complete, ufc_elevation, how="outer", on=["location", "location"])
```

```python
ufc_complete
```

```python
weight_bucket = []
for x in range(len(ufc_complete)):
    if ufc_complete["weight_class"][x] == "Flyweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Bantamweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Featherweight":
        weight_bucket.append("Light")
    elif ufc_complete["weight_class"][x] == "Lightweight":
        weight_bucket.append("Middle")
    elif ufc_complete["weight_class"][x] == "Welterweight":
        weight_bucket.append("Middle")
    elif ufc_complete["weight_class"][x] == "Middleweight":
        weight_bucket.append("Heavy")
    elif ufc_complete["weight_class"][x] == "Light Heavyweight":
        weight_bucket.append("Heavy")
    elif ufc_complete["weight_class"][x] == "Heavyweight":
        weight_bucket.append("Heavy")
```

```python
ufc_complete["weight_buckets"] = weight_bucket
```

```python
blue_win = []
for x in range(len(ufc_complete)):
    if ufc_complete["B_current_win_streak"][x] == 0:
        blue_win.append(0)
    else:
        blue_win.append(1)
```

```python
ufc_complete["b_hot_or_not"] = blue_win
```

```python
red_win = []
for x in range(len(ufc_complete)):
    if ufc_complete["R_current_win_streak"][x] == 0:
        red_win.append(0)
    else:
        red_win.append(1)
```

```python
ufc_complete["r_hot_or_not"] = red_win
```

```python
ufc_complete = pd.get_dummies(ufc_complete, columns=["Winner", "title_bout", 'B_Stance', 'R_Stance', "weight_buckets"])
```

```python
ufc_ml = ufc_complete[[
       'no_of_rounds', 'B_current_lose_streak', 'B_current_win_streak',
       'R_current_lose_streak', 'R_current_win_streak',
       'B_total_rounds_fought', 'R_total_rounds_fought', 'B_wins', 'R_wins',
       'B_Height_cms', 'R_Height_cms', 'B_Reach_cms', 'R_Reach_cms',
        'B_age', 'R_age','elevation (ft)', 'b_hot_or_not',
       'r_hot_or_not', 'Winner_Blue', 'title_bout_False',
       'title_bout_True', 'B_Stance_Open Stance', 'B_Stance_Orthodox',
       'B_Stance_Southpaw', 'B_Stance_Switch', 'R_Stance_Open Stance',
       'R_Stance_Orthodox', 'R_Stance_Southpaw', 'R_Stance_Switch',
       'weight_buckets_Heavy', 'weight_buckets_Light',
       'weight_buckets_Middle']]
```
