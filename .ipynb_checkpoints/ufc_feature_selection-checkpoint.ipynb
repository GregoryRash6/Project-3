```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define File Path
complete = "ufc_feature_selection.csv"

# # Colab URL
# complete = 'https://raw.githubusercontent.com/GregoryRash6/Project-3/master/ufc_feature_selection.csv'

# Read CSV
data = pd.read_csv(complete)
```

# Isabella's Feature: BMI

```python
# Find the Greater BMI
data["greater_bmi"] = data[["B_BMI", "R_BMI"]].max(axis=1)

# Empty List
winner_bmi = []
high_bmi_win = []

# Find BMI of Winner
for x in range(len(data)):
    if data["Winner_Blue"][x] == 0:
        winner_bmi.append(data["R_BMI"][x])
    elif data["Winner_Blue"][x] == 1:
        winner_bmi.append(data["B_BMI"][x])

# Compare Winner BMI to Greater BMI        
for x in range(len(data)):
    if winner_bmi[x] == data["greater_bmi"][x]:
        high_bmi_win.append("True")
    else:
        high_bmi_win.append("False")
        
# Plot New Figure
plt.figure()

# Create Labels
labels = ['Higher BMI Won', 'Lower BMI Won']

# Count True Values
bmi_true = high_bmi_win.count("True")

# Count False Values
bmi_false = high_bmi_win.count("False")

# Get Pie Slices
sizes = [bmi_true, bmi_false]

# Edit Colors
colors = ['gold', 'lightskyblue']

# Plot Pie Graph
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)

# Add Legend
plt.legend(patches, labels, loc="best")

# Add Axis
plt.axis('equal')

# Tight Layout
plt.tight_layout()

# Show Plot
plt.show()
```

# Gregory's Feature: Elevation

```python
# Find Top 15 Most Frequent Locations
loca = data["location"].value_counts()[:15].index

# Set Empty String
ele = []

# Find Top 15 Most Frequent Locations' Elevations
for x in range(len(loca)):
    ele.append(data.loc[data["location"] == loca[x], "elevation (ft)"].iloc[0])

# Plot New Figure
plt.figure()

# Bar Chart
plt.bar(loca, ele, color='lightskyblue')

# Set XTicks
plt.xticks(loca, rotation="vertical")

# Set Title
plt.title("Top 15 Locations")

# Set Y Label
plt.xlabel("Location")

# Set X Label
plt.ylabel("Elevation (ft)")

# Display Chart
plt.show()

# Tight Layout
plt.tight_layout()
```

# Viola's Feature: Attendance

```python
#Dataset with just location and attendacne
new_data = data[['location','attendance ']]

#Only choose 'loca' locations
test = new_data[new_data["location"].isin(loca)]

# Find Top 15 Most Frequent Locations' Average Attendance
attendance_list = []

for x in range(len(loca)):
    attendance_list.append(test[test["location"] == loca[x]]["attendance "].mean())
attendance_list

# Plot New Figure
plt.figure()

# Bar Chart
plt.bar(loca, attendance_list, color='lightskyblue')

# Set XTicks
plt.xticks(loca, rotation="vertical")

# Set YTicks
plt.yticks(np.arange(0, 50000, step=10000))

# Set Title
plt.title("Top 15 Locations")

# Set Y Label
plt.xlabel("Location")

# Set X Label
plt.ylabel("Avg. Attendance")

# Display Chart
plt.show()

# Tight Layout
plt.tight_layout()
```

# Emmarose's Feature: Pollution

```python
# Save Pollution Variables
very_bad = len(data[data["Pollution"] == "Very Bad"])
bad = len(data[data["Pollution"] == "Bad"])
moderate = len(data[data["Pollution"] == "Moderate"])
good = len(data[data["Pollution"] == "Good"])
labels = ["Very Bad", "Bad", "Moderate", "Good"]
values = [very_bad, bad, moderate, good]

# Plot New Figure
plt.figure()

# Bar Chart
plt.bar(labels, values, color='lightskyblue')

# Set XTicks
plt.xticks(labels, rotation="vertical")

# Set Title
plt.title("Number of Matches by Air Quality")

# Set Y Label
plt.xlabel("Air Quality")

# Set X Label
plt.ylabel("Number of Matches")

# Display Chart
plt.show()

# Tight Layout
plt.tight_layout()
```
