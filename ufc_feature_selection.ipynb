```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define File Path
complete = "ufc_feature_selection.csv"

# # Colab URL
# complete = 'https://raw.githubusercontent.com/GregoryRash6/Project-3/master/ufc_feature_selection.csv'

# Read CSV
data = pd.read_csv(complete)

# Show Data
data
```

```python
# Show All Columns
sorted(data.columns)
```

```python
# Find Top 15 Most Frequent Locations
loc = data["location"].value_counts()[:15].index

# Set Empty String
ele = []

# Find Top 15 Most Frequent Locations' Elevations
for x in range(len(loc)):
    ele.append(data.loc[data["location"] == loc[x], "elevation (ft)"].iloc[0])
```

```python
# Plot New Figure
plt.figure()

# Bar Chart
plt.bar(loc, ele)

# Set XTicks
plt.xticks(loc, rotation="vertical")

# Set Title
plt.title("Top 15 Locations")

# Set Y Label
plt.xlabel("Location")

# Set X Label
plt.ylabel("Elevation (ft)")

# Display Chart
plt.show()

# Tight Layout
plt.tight_layout()
```
